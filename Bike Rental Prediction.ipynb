# Outlier Analysis
import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(data):
    
    threshold=0.10
    mean_1 = np.mean(data)
    std_1 =np.std(data)
    
    
    for y in data:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers

# Removing the Outliers!

```python
dailyDataWithoutOutliers = df[np.abs(df["cnt"]-df["cnt"].mean())<=(3*df["cnt"].std())] 
```
