```python
import yfinance as yf
import pandas as pd
from sklearn.model_selection import train_test_split

# Fetch Sun Pharma data using yfinance
ticker = "SUNPHARMA.NS"
data = yf.download(ticker, start="2010-01-01", end="2023-05-11")

# Split the data into training and testing based on index
train_data = data[data.index <= '2018-12-31']
test_data = data[data.index > '2018-12-31']

# Verify the training and testing set sizes
print("Training Data Shape:", train_data.shape)
print("Testing Data Shape:", test_data.shape)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Prepare the training data
train_features = train_data[['Open', 'High', 'Low', 'Close', 'Volume']].values

# Scale the features using MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_features)

# Prepare the target labels
target_labels = ['Open', 'High', 'Low']
train_labels = train_data[target_labels].values

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=64, activation='relu', input_shape=(train_scaled.shape[1], 1)))
model.add(Dense(units=len(target_labels)))
model.compile(optimizer='adam', loss='mse')

# Reshape the input data to match LSTM input shape
train_scaled = train_scaled.reshape(train_scaled.shape[0], train_scaled.shape[1], 1)

# Train the LSTM model
model.fit(train_scaled, train_labels, epochs=50, batch_size=32)

# Prepare the testing data
test_features = test_data[['Open', 'High', 'Low', 'Close', 'Volume']].values

# Scale the testing features using the trained scaler
test_scaled = scaler.transform(test_features)

# Reshape the testing data to match LSTM input shape
test_scaled = test_scaled.reshape(test_scaled.shape[0], test_scaled.shape[1], 1)

# Make predictions on the testing data
predictions = model.predict(test_scaled)

# Inverse transform the scaled predictions to get the actual values
predicted_values = scaler.inverse_transform(predictions)

# Extract the predicted open, high, and low prices
predicted_open = predicted_values[:, 0]
predicted_high = predicted_values[:, 1]
predicted_low = predicted_values[:, 2]

# Print the predicted prices
print("Predicted Open:", predicted_open)
print("Predicted High:", predicted_high)
print("Predicted Low:", predicted_low)
```
